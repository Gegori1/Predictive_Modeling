```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
```

```python
path_to_file = Path().resolve().parent / "Data" / "X_nueva_predicha_035"
data = pd.read_csv(path_to_file)
```

```python
data.Lambda.unique()
```

```python
dn = (
    data
    .query("Lambda == 596.42147")
    .query("`R-predicha` < 10")
    .drop_duplicates()
    )
```

```python
fig = px.scatter_3d(dn, x='espAl2O3', y='espSiO2', z='R-predicha', color='R-predicha', opacity=0.7)
```

```python
fig.show()
```
